# Enunciado

La empresa de Ciberseguridad Seguri-net está analizando dos métodos de encriptación para archivos desde $10$ MB hasta $80$ MB. En el primer método, por cada incremento de $2$ MB en el tamaño del archivo, el tiempo de encriptación se incrementa de manera constante en $0,8$ segundos, tal como muestra la tabla (Método A). Mientras que el segundo método (Método B) tiene un tiempo de encriptación, también en segundos, que varía de acuerdo a un modelo cuadrático.

#### Método A:
<center>

| Tamaño del archivo (MB) | Tiempo de encriptación (segundos) |

|$$10$$                  |$$7,4$$               |

|$$40$$                  |$$19,4$$              |
|$$53$$                  |$$24,6$$              |
|$$76$$                  |$$33,8$$              |

</center>

#### Método B:


$$g(x)=0,01x^2-0,25x+10$$

```python
# --- Pregunta 1 ---
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


#Primero generamos los valores de la variable x y la variable y
x = np.array([10, 22, 40, 53, 76])
y= np.array([7.4, 12.2, 19.4, 24.6, 33.8])
#Descubrimos la pendiente y el coeficiente de posicion.
pendiente, intercepto = np.polyfit(x,y,1)
print("La pendiente es",round(pendiente,2),"y el coeficiente de posición es",round(intercepto,2))

# definimos las funciones A.
def A(X):
    return X*0.4 + 3.4
# Respuesta
print("La funcion con el metodo A es de tipo lineal y su forma algebraica seria x*0,4 + 3,4")
```

```python
# --- Pregunta 2 ---

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def A(X):
    return X*0.4 + 3.4


def B(X):
    return 0.01*X**2 - 0.25*X+ 10


#Cuando el tamaño es de 27 mb, cuanto tiempo tarda?

tiempo_A = 27*0.4+3.4


tiempo_B = 0.01*27**2 - 0.25*27 + 10

print(f"Los tiempos son: forma A:  {tiempo_A :.1f} y el tiempo de la forma B: {tiempo_B :.1f} ")

if tiempo_A < tiempo_B:
    print(f"La forma A se demora menos con un tiempo de {tiempo_A} segundos")
else:
    print(f"La forma B se demora menos con un tiempo de {tiempo_B} segundos")

```

```python
# --- Pregunta 3 ---

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
def A(X):
    return X*0.4 + 3.4


def B(X):
    return 0.01*X**2 - 0.25*X + 10




# La función ahora acepta el argumento 'x'
def def_interseccion(x):
    return A(x) - B(x)

# fsolve encontrará los valores de 'x' donde A(x) - B(x) = 0
sol = fsolve(def_interseccion, [12, 52])

x1, x2 = sorted(sol)

print(f'Los puntos de interseccion son {x1 :.2F}, {x2 :.2F}')



```

```python
# --- Pregunta 4 ---


import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def A(X):
    return X*0.4 + 3.4


def B(X):
    return 0.01*X**2 - 0.25*X + 10







X = np.linspace(0,80,600)
plt.figure(figsize=(10,7))

plt.plot(X,A(X), label ="Metodo A")
plt.plot(X,B(X), label ="Metodo B")

#Para configurar ambos ejes:
plt.axis([0,80,0,70]) #Si desea configurarlos de a uno, usar plt.xlim() y plt.ylim()
plt.title('Metodos de encriptacion')
plt.xlabel('Tamaño de archivo en mb')
plt.ylabel('Tiempo de encriptacion en segundos')
plt.grid(True)
plt.legend()
plt.show()


#Buscar el rango en que el metodo B toma menos tiempo

print("Segun el grafico podemos estimar que entre 12mb y 52mb aproximadamente el metodo B es es mas conveniente")

```
