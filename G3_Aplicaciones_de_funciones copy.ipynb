# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

1.Primero generamos las funciones
2.Creamos el grafico y graficamos.


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def r(x):
    return -1.5*x**2+11.5*x-15

def g(x):
    return 1.58*x**4 - 19.17*x**3 + 80.92*x**2 - 139.33*x + 85



# --- Parte 1: Graficar ambas funciones ---

# 3. Crear el dominio: 500 puntos para 'x' entre 0 y 6 para una curva suave.
x_valores = np.linspace(0, 6, 900)

# 4. Calcular los valores de 'y' para cada función usando los puntos de 'x'
y_valores_r = r(x_valores)
y_valores_g = g(x_valores)

# 5. Configurar el gráfico
plt.figure(figsize=(7, 4)) # Define un buen tamaño para el gráfico

# 6. Dibujar cada función en el gráfico
plt.plot(x_valores, y_valores_r, label='r(x)', color='blue', linewidth=2)
plt.plot(x_valores, y_valores_g, label='g(x)', color='red', linewidth=2)

# 7. Aplicar los límites para los ejes X e Y, como lo pide el problema
plt.xlim(0, 6)
plt.ylim(-10, 15)

# 8. Añadir elementos para que el gráfico sea más claro
plt.xlabel("Eje X", fontsize=12)
plt.ylabel("Eje Y", fontsize=12)
plt.title("Gráfico de r(x) y g(x)", fontsize=14)
plt.grid(True) # Pone una grilla de fondo
plt.legend()   # Muestra las etiquetas de cada línea (labels)
plt.axhline(0, color='black', linewidth=0.5) # Dibuja el eje X
plt.axvline(0, color='black', linewidth=0.5) # Dibuja el eje Y

# 9. Mostrar el gráfico final
plt.show()


# --- Parte 2: Encontrar los puntos de intersección ---

# 10. Para encontrar donde r(x) = g(x), buscamos las raíces de la función r(x) - g(x) = 0
def diferencia(x):
    return r(x) - g(x)

# 11. Mirando el gráfico, podemos estimar que las intersecciones están cerca de x=2 y x=5.1.
#     Usaremos estos valores como "puntos de partida" para que fsolve encuentre la solución exacta.
x_interseccion1 = fsolve(diferencia, 2)
x_interseccion2 = fsolve(diferencia, 5.08)

# 12. Para obtener el punto completo (x, y), evaluamos una de las funciones originales
#     en los puntos 'x' que encontramos. Usaremos r(x) por ser más simple.
y_interseccion1 = r(x_interseccion1)
y_interseccion2 = r(x_interseccion2)

# 13. Imprimir los resultados de forma clara
#     El [0] se usa porque fsolve devuelve un arreglo, y queremos solo el primer elemento.
#     El :.4f formatea el número para que muestre solo 4 decimales.
print("--- Puntos de Intersección ---")
print(f"La primera intersección es en el punto ({x_interseccion1[0]:.4f}, {y_interseccion1[0]:.4f})")
print(f"La segunda intersección es en el punto ({x_interseccion2[0]:.4f}, {y_interseccion2[0]:.4f})")



```

# Problema 2

Sean las funciones:

\begin{split}
h(x) & = 2(x - 2)^{2}\\
i(x) & = x + 1
\end{split}

<br>

1. Grafique ambas funciones en un mismo gráfico utilizando *Matplotlib*.
2. ¿En qué punto(s) interseca la función $h(x)$ al eje $x$? ¿y la función $i(x)$?
3. ¿Las gráficas de las funciones se intersecan en algún punto?
4. ¿Para qué valores de $x$ ocurre que $h(x) \leq i(x)$?
5. Determine el punto en que $h(x) = 50$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def h(x):
    return 2 * (x - 2)**2


def i(x):
    return x + 1


# --- Parte 1: Graficar ambas funciones ---

# Crear un rango de valores de 'x' para tener un buen gráfico.
# Elegimos de -3 a 6 para ver bien todos los puntos de interés.
x_vals = np.linspace(-3, 6, 400)

# Calcular los valores de 'y' para cada función
y_h = h(x_vals)
y_i = i(x_vals)

# Configurar y mostrar el gráfico
plt.figure(figsize=(10, 7))
plt.plot(x_vals, y_h, label='h(x) = 2(x-2)²', color='purple')
plt.plot(x_vals, y_i, label='i(x) = x+1', color='green')

# Añadir etiquetas y títulos para claridad
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico de h(x) vs i(x)")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

# --- Parte 2: Intersección con el eje X ---

# La intersección con el eje X ocurre cuando y=0.
# Para h(x): 2(x-2)² = 0  => (x-2)² = 0 => x-2=0 => x=2
# Para i(x): x+1 = 0 => x=-1
print("--- Intersecciones con el Eje X ---")
print("La función h(x) interseca el eje X en el punto (2, 0).")
print("La función i(x) interseca el eje X en el punto (-1, 0).")
print("-" * 35) # Separador
# --- Parte 3: Intersección entre las funciones ---

# Para encontrar donde h(x) = i(x), creamos la función diferencia.
def diferencia_h_i(x):
    return h(x) - i(x)

# Mirando el gráfico, estimamos que las intersecciones están cerca de x=1 y x=3.5
x_intersec1 = fsolve(diferencia_h_i, 1)
x_intersec2 = fsolve(diferencia_h_i, 3.5)

# Calculamos las coordenadas 'y' para los puntos de intersección
y_intersec1 = i(x_intersec1) # Usamos i(x) por ser más simple
y_intersec2 = i(x_intersec2)

print("--- Intersecciones entre h(x) e i(x) ---")
print("Sí, las gráficas de las funciones se intersecan en dos puntos:")
print(f"Punto 1: ({x_intersec1[0]:.3f}, {y_intersec1[0]:.3f})")
print(f"Punto 2: ({x_intersec2[0]:.3f}, {y_intersec2[0]:.3f})")
print("-" * 35) # Separador

# --- Parte 4: ¿Cuándo h(x) <= i(x)? ---

# Mirando el gráfico, la parábola (morada) está por debajo de la recta (verde)
# entre los dos puntos de intersección que acabamos de encontrar.
print("--- Intervalo donde h(x) <= i(x) ---")
print("La condición h(x) <= i(x) se cumple en el intervalo entre los puntos de intersección.")
print(f"Es decir, para x en el intervalo [{x_intersec1[0]:.3f}, {x_intersec2[0]:.3f}]")
print("-" * 35) # Separador

# --- Parte 5: Determinar el punto en que h(x) = 50 ---

# Resolvemos la ecuación h(x) - 50 = 0.
def h_menos_50(x):
    return h(x) - 50

# Matemáticamente: 2(x-2)²=50 => (x-2)²=25 => x-2 = ±5
# Solución 1: x-2 = 5  => x=7
# Solución 2: x-2 = -5 => x=-3
# Usamos fsolve para encontrar las soluciones numéricamente
sol1 = fsolve(h_menos_50, 5)  # Estimación positiva
sol2 = fsolve(h_menos_50, -1) # Estimación negativa

print("--- Puntos donde h(x) = 50 ---")
print(f"La función h(x) es igual a 50 cuando x = {sol1[0]:.3f} y cuando x = {sol2[0]:.3f}.")
```

# Problema 3

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


#Ojo, el problema no indica unidad de medida, así que asumimos que son unidades genéricas.

# --- Parte 1: Graficar ambas funciones ---

# 1. Definir las funciones
def w(x):
    return 0.6*x**2+1.5
def t(x):
    return 0.8*x**3-10*x+10 

# 3. Crear el dominio: 700 puntos para 'x' entre -3 y 4 para una curva suave.
x_valores = np.linspace(-3, 4, 700)

# 4. Calcular los valores de 'y' para cada función usando los puntos de 'x'
y_valores_w = w(x_valores)
y_valores_t = t(x_valores)

# 5. Configurar el gráfico
plt.figure(figsize=(7, 3)) # Define un buen tamaño para el gráfico

# 6. Dibujar cada función en el gráfico
plt.plot(x_valores, y_valores_w, label='w(x)', color='blue', linewidth=2)
plt.plot(x_valores, y_valores_t, label='t(x)', color='red', linewidth=2)

# 7. Aplicar los límites para los ejes X e Y, como lo pide el problema
plt.xlim(-3, 4)
plt.ylim(-4, 24)

# 8. Añadir elementos para que el gráfico sea más claro
plt.xlabel("Eje X", fontsize=12)
plt.ylabel("Eje Y", fontsize=12)
plt.title("Gráfico de w(x) y t(x)", fontsize=14)
plt.grid(True) # Pone una grilla de fondo
plt.legend()   # Muestra las etiquetas de cada línea (labels)
plt.axhline(0, color='black', linewidth=0.5) # Dibuja el eje X
plt.axvline(0, color='black', linewidth=0.5) # Dibuja el eje Y

# 9. Mostrar el gráfico final
plt.show()

# --- Parte 2: Encontra intervalos en que w(x) >=  t(x) ---
# 10. Para encontrar donde w(x) = t(x), buscamos las raíces de la función w(x) - t(x) = 0
def diferencia_w_t(x):
    return w(x) - t(x)

# 11. Mirando el gráfico, podemos estimar que las intersecciones están cerca de x=0.8 y x=3.3.
x_interseccion1 = fsolve(diferencia_w_t, 0.8)
x_interseccion2 = fsolve(diferencia_w_t, 3.3)   
# 12. Para obtener el punto completo (x, y), evaluamos una de las funciones originales
#     en los puntos 'x' que encontramos. Usaremos w(x) por ser más simple.
y_interseccion1 = w(x_interseccion1)
y_interseccion2 = w(x_interseccion2)
# 13. Imprimir los resultados de forma clara
print("--- Puntos de Intersección ---")
print(f"La primera intersección es en el punto ({x_interseccion1[0]:.4f}, {y_interseccion1[0]:.4f})")
print(f"La segunda intersección es en el punto ({x_interseccion2[0]:.4f}, {y_interseccion2[0]:.4f})")

print("El intervalo donde w(x) >= t(x) es entre los puntos de intersección:")
print(f"Es decir, para x en el intervalo [{x_interseccion1[0]:.4f}, {x_interseccion2[0]:.4f}]")
print("-" * 35) # Separador



# --- Parte 3: Encontrar los puntos tal que t(x)=6 ---
```
